**Bibliotecas**

```
import pandas as pd
```

**Dados**

```
pernambuco = pd.read_csv('https://raw.githubusercontent.com/tiagopessoalima/covid-19/master/dados/pernambuco.csv',sep=';')
obito = pernambuco[(pernambuco.evolucao=='OBITO') & (pernambuco.classe=='CONFIRMADO')]
```

**Comorbidades**

```
comorbidades = obito.groupby('comorbidades').size()
comorbidades.sort_values(inplace=True)
```

```
comorbidades[comorbidades>1].plot.bar(figsize=(30,5))
```

```
comorbidades[comorbidades>5].plot.bar(figsize=(30,5))
```

```
comorbidades[comorbidades>10].plot.bar(figsize=(30,5))
```

```
comorbidades[comorbidades>20].plot.bar(figsize=(30,5))
```

```
comorbidades[comorbidades>30].plot.bar(figsize=(30,5))
```

**Data do Óbito**

```
dt_obito = obito.copy()
dt_obito['dt_obito']=pd.to_datetime(dt_obito['dt_obito'])
dt_obito['Semana'] = dt_obito['dt_obito'].dt.to_period('W-THU')
dt_obito.groupby(['Semana', 'faixa_etaria']).size().unstack(0).plot.bar(figsize=(20,5))
```

```
dt_obito.groupby(['Semana', 'faixa_etaria']).size().unstack(1).plot.bar(figsize=(20,5))
```

**Tempo**

```
tempo = obito.copy()
tempo['dt_obito']=pd.to_datetime(tempo['dt_obito'])
tempo['dt_primeiros_sintomas']=pd.to_datetime(tempo['dt_primeiros_sintomas'])
tempo['dias'] = abs(tempo['dt_obito'] - tempo['dt_primeiros_sintomas'])
```

```
dias = tempo.groupby('dias')
dias.size().plot.bar(figsize=(30,5))
```

```
dias_sexo = tempo.groupby(['dias','Sexo'])
dias_sexo.size().unstack(1).plot.bar(figsize=(20,10))
```

```
dias_hospitalizacao = tempo.groupby(['dias','hospitalizacao'])
dias_hospitalizacao.size().unstack(1).plot.bar(figsize=(20,10))
```
